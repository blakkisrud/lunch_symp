```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
def bix_exp(time_vec, a1, a2, b1, b2):
    
    alpha = a1*np.exp(-b1*time_vec)
    beta = a2*np.exp(-b2*time_vec)
    
    return alpha + beta
```

```python

sns.set_style('white')

plt.figure()
x = np.array([1,24,24*2,24*5,24*7])
x_full = np.arange(180)
print x
y = bix_exp(x,3,-3,0.01,0.9)
y_full = bix_exp(x_full, 3, -3, 0.01, 0.9)
print y

for i in range(1,len(x)+1):
    plt.plot(x[0:i],y[0:i],'o', markersize=8, color = 'g')
    plt.ylim([0, 3.1])
    plt.xlim([0, 180])
    plt.ylabel('Activity (MBq)')
    plt.xlabel('Time (hrs)')
    plt.savefig(str(i) + 'dot.png')

plt.plot(x_full, y_full, '--')
plt.savefig('stripe.png')

plt.fill_between(x_full, 0, y_full)
plt.savefig('filled.png')
```
